```python
# 1 Birthday Cake Candles
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the birthdayCakeCandles function below.
def birthdayCakeCandles(ar):
    k=0
    a = max(ar)
    for i in ar:
        if i==a:
            k=k+1
    return k

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    ar_count = int(input())

    ar = list(map(int, input().rstrip().split()))

    result = birthdayCakeCandles(ar)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
# 2 Kangaroo
#!/bin/python3
import math
import os
import random
import re
import sys
# Complete the kangaroo function below.
def kangaroo(x1, v1, x2, v2):
    if(x1 != x2 and v1==v2):
        return('NO')
    k = (x1-x2)/(v2-v1) # just solved the equation x1+kv1=x=x2+kv2 and then checked if it's a positive int
    if(k>=0 and k==int(k)):
        return("YES")
    else:
        return('NO')

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    x1V1X2V2 = input().split()

    x1 = int(x1V1X2V2[0])

    v1 = int(x1V1X2V2[1])

    x2 = int(x1V1X2V2[2])

    v2 = int(x1V1X2V2[3])

    result = kangaroo(x1, v1, x2, v2)

    fptr.write(result + '\n')

    fptr.close()
```

```python
# 4 Recursive Digit Sum
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the superDigit function below.
def superDigit(n, k):
    a =  sum([int(i)*k for i in str(n)])
    if a<=9:
        return a
    else:
        return(superDigit(a,1))#yes putting this 1 is like "cheating" but for the purpose of optimization, being the number not so big, it's ok.
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    nk = input().split()

    n = nk[0]

    k = int(nk[1])

    result = superDigit(n, k)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
# 5 Insertion Sort - Part 1
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the insertionSort1 function below.
def insertionSort1(n, arr):
        n=n-1
        key = arr[n]
        j = n-1
        while(j>=0 and arr[j]>key):
            arr[j+1]=arr[j]
            print(*arr, sep = ' ')
            j=j-1
        arr[j+1]=key   
        print(*arr, sep = ' ')
if __name__ == '__main__':
    n = int(input())

    arr = list(map(int, input().rstrip().split()))

    insertionSort1(n, arr)
```
